```python
import torch
import torch.nn as nn
```

loss = F(y, 天) = F(ground_truth, prediction) -> float

y : an int (class label)

天 : a tensor of size number of classes

to compare y and 天 we should transform y into a tensor this could be done using torch

```python
import torch.nn.functional as F

F.one_hot(torch.tensor(0), num_classes=3)
```

once this is done, we can pass the two input value to loss function to calculate the loss value

```python
from torch.nn import CrossEntropyLoss

scores = torch.tensor([[0.2, -0.3]])
one_hot_target = torch.tensor([[1, 0]])

criterion = CrossEntropyLoss()
loss = criterion(scores.double(), one_hot_target.double()) #F(y, 天)

print(loss)
```
