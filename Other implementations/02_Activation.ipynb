```python
import torch
import torch.nn as nn
```

```python
# Use Sigmoid for binary classification 0/1

X = torch.tensor([[6.0]])

sigmoid = nn. Sigmoid()

Y = sigmoid(X)
print(Y)
```

```python
# Use Softmax for multi-class classification

X = torch.tensor([[4.3, 6.1, 2.3]])

probas = nn.Softmax(dim=-1)

Y = probas(X)
print(Y)
```
