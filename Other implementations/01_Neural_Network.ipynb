```python
import torch
```

```python
#Input tensor of 3 features

X = torch.tensor([0.3474, 0.4547, -0.2356])

#X : input tensor

print(X.dtype)
print(X.shape)
```

```python
#Linear : W et b
linear_layer = nn.Linear(in_features=3, out_features=2)

#Y : output tensor (sa dimension est out = 2)
Y = linear_layer(X)

print(Y)
```

```python
#Si par exemple X est une matrice de m lignes (m donnees) y comportera m lignes

data = torch.randn(7, 3)
X = torch.tensor(data)

print(X, '\n')

linear_layer = nn.Linear(in_features=3, out_features=2)
Y = linear_layer(X)

print(Y)
```

```python
#Affichage des W et b
print(linear_layer.weight.shape)
print(linear_layer.bias.shape)
```

```python
print(Y, '\n\n')

print(X @ linear_layer.weight.T + linear_layer.bias)
```

```python
#Multiple Layers
#Network creation

model = nn.Sequential(
    nn.Linear(10, 18),
    nn.Linear(18, 20),
    nn.Linear(20, 5)
)

data = torch.randn(1, 10)
X = torch.tensor(data)

print(X)

#Pass X into the model
Y = model(X)

print(Y.shape)
print(Y)
```
