```python
import torch
import torch.nn as nn
```

```python
n_features=12
n_classes=3

model = nn.Sequential(nn.Linear(n_features, 8),
                      nn.Linear(8, 4),
                      nn.Linear(4, n_classes))
```

Model's capacity : Number of parameters

.numel() returns the number of elements in the tensor, therefore we iterate each parameter and add the number of parameters to to total number

```python
total = 0
for p in model.parameters():
  total+= p.numel()
print(total)
```
